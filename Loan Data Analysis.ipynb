# Binning

```python
def df_sanct(x):
    if x <= 300000:
        return "Sanctioned amount 300,000 or less"
    else:
        return "Sanctioned amount over 300,000"
```

```python
def df_disbursed(x):
    if x <= 300000:
        return "Disbursed amount 300,000 or less"
    else:
        return "Disbursed amount more than 300,000"
```

```python
def df_sec_sanct(x):
    if x <= 10000:
        return "Secondary Sanctioned amount less than 10k"
    else:
        return "Secondary Sanctioned amount more than 10k"
```

```python
def df_sec_disbursed(x):
    if x <= 10000:
        return "Secondary disbursed amount under 10k"
    else:
        return "Secondary disbursed amount over 10k"
```

```python
def df_BS(x):
    if x <=300:
        return "Bureau score less than 300"
    elif x >=301 and x <= 500:
        return "Bureau score of 301 to 500"
    else:
        return "Bureau score more than 500"
```

```python
def df_custid(x):
    if x >= 400000 and x<500000:
        return "customer ID: 400k to 500k"
    elif x >=500000 and x<=600000:
        return "customer ID: 500k to 600k"
    else:
        return "customer ID: greater than 600k"
```

```python
def df_asset_cost(x):
    if x <= 50000:
        return "Asset cost below 50000"
    elif x >50000 and x<=100000:
        return "Asset cost 50000 to 100000"
    else:
        return "Asset cost greater than 100000"
```

```python
def df_new_loan(x):
    if x <= 0:
        return "0 new loans last 6 months"
    else:
        return "1 or more loans in last 6 months"
```

```python
def df_pri_loan(x):
    if x <= 3:
        return "3 or less total loans"
    else:
        return "4 or more total loans"
```

```python
def df_act_loan(x):
    if x <= 1:
        return "1 or less active loans"
    else:
        return "2 or more active loans"
```

```python
def df_over_loan(x):
    if x <= 0:
        return "0 overdue loans"
    else:
        return "1 or more overdue loans"
```

```python
def df_pri_outstanding(x):
    if x < 1:
        return "Negative outstanding balance"
    elif x > 1 and x <= 300000:
        return "Outstanding balance between 1 to 300,000"
    else:
        return "Outstanding balance more than 300,000"
```

```python
def df_sec_out(x):
    if x < 5000:
        return "Secondary balance 5,000 or less"
    elif x > 5000 and x <= 50000:
        return "Secondary balance between 5,000 to 50,000"
    else:
        return "Secondary balance over 50,000"
```

# Logistic regression analysis

```python
X = loan_clean.drop(['loan_default'],axis=1)
X.head()
Y = loan_clean['loan_default']
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, Y, test_size=0.3, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
props = classifier.predict_proba(X_test)
print (props)
```

```python
from sklearn import metrics

print (metrics.accuracy_score(y_test, y_pred))
print (metrics.roc_auc_score(y_test, props[:, 1]))
```

```python
print (metrics.confusion_matrix(y_test, y_pred))
```
